```python
def f(x): return x**3 + 2*x**2 - x -2
def df(x): return 3*x**2  + 4*x - 1
```

```python


x0 = 0.05
tol = 1E-6
NumIterations=20

x=x0
for i in range(NumIterations):
    fx = f(x)
    if abs(fx) < tol:
        print(f"Root found: {x} (after {i} iterations)")
        break
    dfx = df(x)
    if abs(dfx) < tol:
        print(f"Derivative is zero at {x}: can't continue")
    x_new = x - fx/dfx
    x = x_new
else:
    print(f"Max iterations reached. Last value: x={x} f(x)={f(x)}")
    
```

```python
def newt(f,df,x0,tol=1E-6,NumIterations=20):
    x=x0
    for i in range(NumIterations):
        fx = f(x)
        if abs(fx) < tol:
            print(f"Root found: {x} (after {i} iterations)")
            break
        dfx = df(x)
        if abs(dfx) < tol:
            print(f"Derivative is zero at {x}: can't continue")
        x_new = x - fx/dfx
        x = x_new
    else:
        print(f"Max iterations reached. Last value: x={x} f(x)={f(x)}")
```

```python
newt(f,df,0)
```

```python
newt(f,df,5)
```

```python
import numpy as np
for x0 in np.linspace(-3,3):
    newt(f,df,x0)
```
