```python
ship_costs = { ("Santa Fe", "Seattle") : 5, 
               ("Santa Fe", "LA"): 4,
               ("Santa Fe", "NY"): None,
               ("Santa Fe", "Dallas"): 3,
               ("Santa Fe", "Atlanta"): 8,
               ("El Paso", "Seattle") : 5, 
               ("El Paso", "LA"): 4,
               ("El Paso", "NY"): None,
               ("El Paso", "Dallas"): 3,
               ("El Paso", "Atlanta"): 8
             }

#transfer_costs { (): }

warehouses = [ "Santa Fe", "El Paso" ]
stores = ["Seattle", "LA", "NY", "Dallas", "Atlanta"]
```

```python
for wh in warehouses:
    for store in stores:
        cost = ship_costs[(wh,store)]
        if cost:
            print(f"cost to ship from {wh} to {store}: {cost}")
        else: 
            print(f"no route")
```

```python
for s1 in stores:
    for s2 in stores:
        print(f"{s1} -> {s2}")
```
